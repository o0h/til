### DataFrame.append(DataFrame)で結合

```python
import pandas as pd

df = pd.DataFrame({
    'name': ['apple', 'banana', 'strawberry'],
    'color': ['red', 'yellow', 'pink'],
    'unit': [1, 5, 15]
},
    index=['ap', 'bn', 'st']
)

df2 = pd.DataFrame({'name': 'grape', 'color': 'purple', 'unit': 1}, index=['gr'])
df.append(df2)
```

辞書にデータを追加するようなノリで列を追加  
-> `df[新しい列名] = [縦方向の値リスト]` でOK

```python
df['available'] = [True, False, True]
df
```

(ちなみに)集計列の追加も簡単。  
-> `df['新しい列名'] = 集計処理`  
※ 各行において、列に対するmap/reduce処理が走る！みたいなイメージで良さそう

```python
df['is_multi'] = df['unit'] > 1
df
```

```python
df['10unit'] = df['unit'] * 10
df
```
